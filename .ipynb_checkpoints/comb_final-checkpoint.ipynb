```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, cross_validate
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis, QuadraticDiscriminantAnalysis
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.neural_network import MLPClassifier
from Bio.SeqUtils.ProtParam import ProteinAnalysis
from lime import lime_tabular
from IPython.display import display, HTML
import sys
import time
from sklearn.base import clone
import warnings
warnings.filterwarnings('ignore')
```

**Loading the datasets containing all the peptides and one containing all the antimicrobial peptides**

```python

posDF = pd.read_csv(r"C:\Users\Advik Narendran\Downloads\bio project\PositiveSequences.csv")
allDataDF = pd.read_csv(r'C:\Users\Advik Narendran\Downloads\bio project\nlp_data.csv')
```

**Extracting all the peptides which are non antimicrobial in nature**

```python
allDataDF = allDataDF.set_index(allDataDF['DADP ID'])
posDF = posDF.set_index(posDF['DADP ID'])
posID_List = posDF["DADP ID"].tolist()
negDF = allDataDF.drop(posID_List)

invalid_chars = "/"
negDF = negDF[~negDF['Bioactive sequence'].str.contains(invalid_chars)]
posDF = posDF[~posDF['Bioactive sequence'].str.contains(invalid_chars)]

display(HTML('<h2><b>Dataframe with all peptides:</b></h2>'))
print(allDataDF)
display(HTML('<h2><b>Dataframe with only Antimicrobial peptides:</b></h2>'))
print(posDF)
display(HTML('<h2><b>Dataframe with only non Antimicrobial peptides:</b></h2>'))
print(negDF)
```

**Preparing a dataset for the classification using physicochemical properties by obtaining the properties of the peptides using biopython**

```python

def calculate_physicochemical_properties(sequence):
    analyzed_seq = ProteinAnalysis(sequence)
    properties = {
        'MolecularWeight': analyzed_seq.molecular_weight(),
        'IsoelectricPoint': analyzed_seq.isoelectric_point(),
        'Aromaticity': analyzed_seq.aromaticity(),
        'InstabilityIndex': analyzed_seq.instability_index(),
        'Hydrophobicity': analyzed_seq.gravy(),
        'ChargeAtPH7': analyzed_seq.charge_at_pH(7.0),        
    }
    return properties

posDF['Properties'] = posDF['Bioactive sequence'].apply(calculate_physicochemical_properties)
negDF['Properties'] = negDF['Bioactive sequence'].apply(calculate_physicochemical_properties)

pos_properties = pd.DataFrame(posDF['Properties'].tolist(), index=posDF.index)
neg_properties = pd.DataFrame(negDF['Properties'].tolist(), index=negDF.index)
pos_properties['Value'] = 1
neg_properties['Value'] = 0

all_properties = pd.concat([pos_properties, neg_properties])
display(HTML('<h2><b>All the peptides along with their physicochemical properties</b></h2>'))
print(all_properties)
```

**Splitting the datsaset values into train and test for training model**

```python

X1 = all_properties.drop('Value', axis=1)
y1 = all_properties['Value']

X1_train, X1_test, y1_train, y1_test = train_test_split(X1, y1, test_size=0.2, random_state=0)

scaler1 = StandardScaler()
X1_train = scaler1.fit_transform(X1_train)
X1_test = scaler1.transform(X1_test)

```

**Preprocessing the peptides in the dataset for Onehot Encoding**

```python

allDataDF = allDataDF.reset_index(drop=True)
posDF = posDF.reset_index(drop=True)
negDF = negDF.reset_index(drop=True)

posDF['Bioactive sequence'] = (posDF['Bioactive sequence'].str.upper())
negDF['Bioactive sequence'] = (negDF['Bioactive sequence'].str.upper())



posDF['TruncSequence'] = (posDF['Bioactive sequence'].str.slice(0,22))
negDF['TruncSequence'] = (negDF['Bioactive sequence'].str.slice(0,22))


posDF['ModSequence'] = (posDF['TruncSequence'].str.pad(22,"right","#"))
negDF['ModSequence'] = (negDF['TruncSequence'].str.pad(22,"right","#"))


posDF['Value'] = 1
negDF['Value'] = 0

display(HTML('<h2><b>The negative sequences, with preprocessed sequence strings and value columns appended:</b></h2>'))
print(negDF.head(3))


display(HTML('<h2><b>The positive sequences, with preprocessed sequence strings and value columns appended:</b></h2>'))
print(posDF.head(3))

allDataDF2 = pd.concat([posDF, negDF])
display(HTML('<h2><b>The Complete dataset for One Hot Encoding:</b></h2>'))
display(allDataDF2)
allDataDF2 = allDataDF2[['ModSequence', 'Value']]
```

**Completing the dataset by adding the embedded peptide**

```python
alphabet = 'ABCDEFGHIJKLMNOPQRSTUVWXYZ#/ '
char_to_int = dict((c, i) for i, c in enumerate(alphabet))

def one_hot_encode_sequence(sequence):
    integer_encoded = [char_to_int[char] for char in sequence]
    onehot_encoded = list()
    for value in integer_encoded:
        letter = [0 for _ in range(len(alphabet))]
        letter[value] = 1
        onehot_encoded.append(letter)
    return [item for sublist in onehot_encoded for item in sublist]

allDataDF2['EncodedX'] = allDataDF2['ModSequence'].apply(one_hot_encode_sequence)


display(HTML('<h2><b>Encoded sequences from the DataFrame:</b></h2>'))
for i in range(3):
    print(f"Sequence {i+1}: {allDataDF2['ModSequence'].iloc[i]}")
    print(f"One-hot Encoded: {allDataDF2['EncodedX'].iloc[i]}")
```

**Splitting the datsaset values into train and test for training model**

```python

X2 = np.array(list(allDataDF2['EncodedX']))
y2 = allDataDF2['Value']

X2_train, X2_test, y2_train, y2_test = train_test_split(X2, y2, test_size=0.2, random_state=0)
```

**A ditionary containing the models that will be used for classification using Physicochemical properties**

```python

models1 = {
    "Logistic Regression": LogisticRegression(C=0.01, max_iter=1000, penalty='l1', solver='saga'),
    "Decision Tree": DecisionTreeClassifier(criterion='entropy', max_depth=50, splitter='random'),
    "SVC": SVC(C=100, gamma='auto', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=500),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(shrinkage='auto', solver='lsqr'),
    "Quadratic Discriminant Analysis": QuadraticDiscriminantAnalysis(),
    "K-Nearest Neighbors": KNeighborsClassifier(metric='manhattan', n_neighbors=3, weights='distance'),
    "Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(activation='tanh', hidden_layer_sizes=(50, 100, 50),learning_rate='adaptive', max_iter=1000)
}
```

**A ditionary containing the models that will be used for classification using One-hot encoding**

```python

models2 = {
    "Logistic Regression": LogisticRegression(C=0.01, max_iter=1000, penalty='l1', solver='saga'),
    "Decision Tree": DecisionTreeClassifier(criterion='entropy', max_depth=50, splitter='random'),
    "SVC": SVC(C=100, gamma='auto', probability=True),
    "Random Forest": RandomForestClassifier(n_estimators=500),
    "Linear Discriminant Analysis": LinearDiscriminantAnalysis(shrinkage='auto', solver='lsqr'),
    "Quadratic Discriminant Analysis": QuadraticDiscriminantAnalysis(),
    "K-Nearest Neighbors": KNeighborsClassifier(metric='manhattan', n_neighbors=3, weights='distance'),
    "Bayes": GaussianNB(),
    "Neural Network": MLPClassifier(activation='tanh', hidden_layer_sizes=(50, 100, 50),learning_rate='adaptive', max_iter=1000)
}
```

**Function to estimate the size the model takes**

```python
def estimate_size(obj):
    return sys.getsizeof(obj) / (1024 * 1024)  
```

**function to train the models for each type of classification**

```python

def train_and_evaluate_models(X_train, y_train, X_test, y_test, models):
    results = []
    trained_models = {}
    
    for model_name, model in models.items():
        start_time = time.time()
        cloned_model = clone(model)  
        cloned_model.fit(X_train, y_train)
        end_time = time.time()
        
        
        execution_time = end_time - start_time
        
        
        model_size = estimate_size(cloned_model)
        scoring = ['accuracy', 'precision_macro', 'recall_macro', 'f1_weighted', 'roc_auc']
        scores = cross_validate(cloned_model, X_train, y_train, scoring=scoring, cv=5, return_train_score=True)

        fit_time = scores['fit_time'].mean()
        score_time = scores['score_time'].mean()
        accuracy = scores['test_accuracy'].mean()
        precision = scores['test_precision_macro'].mean()
        recall = scores['test_recall_macro'].mean()
        f1 = scores['test_f1_weighted'].mean()
        roc_auc = scores['test_roc_auc'].mean()

        model.fit(X_train, y_train)
        
        results.append({
            'Model': model_name,
            'Fitting time': fit_time,
            'Scoring time': score_time,
            'Accuracy': accuracy,
            'Precision': precision,
            'Recall': recall,
            'F1_score': f1,
            'AUC_ROC': roc_auc,
            'Execution Time (s)': execution_time,
            'Model Size (MB)': model_size,
        })

        trained_models[model_name] = model

    return pd.DataFrame(results).sort_values(by='Accuracy', ascending=False), trained_models
```

**Training the models for Physicochemical Properties**

```python

results_method1, trained_models_method1 = train_and_evaluate_models(X1_train, y1_train, X1_test, y1_test, models1)
display(HTML('<h2><b>Model performance for Method 1 (Physicochemical Properties):</b></h2>'))
display(results_method1)
```

**Training the models for One Hot Encoding method**

```python

results_method2, trained_models_method2 = train_and_evaluate_models(X2_train, y2_train, X2_test, y2_test, models2)
display(HTML('<h2><b>Model performance for Method 2 (One-hot Encoding):</b></h2>'))
display(results_method2)
```

**Comparing the two methods using a graphical display**

```python
combined_results = pd.concat([results_method1.assign(Method='Physicochemical Properties'), results_method2.assign(Method='One-hot Encoding')])
fig, axes = plt.subplots(4, 2, figsize=(18, 18))

sns.barplot(x='Accuracy', y='Model', hue='Method', data=combined_results, ax=axes[0, 0])
axes[0, 0].set_title('Model Accuracy')

sns.barplot(x='Precision', y='Model', hue='Method', data=combined_results, ax=axes[0, 1])
axes[0, 1].set_title('Model Precision')

sns.barplot(x='Recall', y='Model', hue='Method', data=combined_results, ax=axes[1, 0])
axes[1, 0].set_title('Model Recall')

sns.barplot(x='F1_score', y='Model', hue='Method', data=combined_results, ax=axes[1, 1])
axes[1, 1].set_title('Model F1 Score')

sns.barplot(x='AUC_ROC', y='Model', hue='Method', data=combined_results, ax=axes[2, 0])
axes[2, 0].set_title('Model ROC AUC')

sns.barplot(x='Fitting time', y='Model', hue='Method', data=combined_results, ax=axes[2, 1])
axes[2, 1].set_title('Model Fitting Time')

sns.barplot(x='Model Size (MB)', y='Model', hue='Method', data=combined_results, ax=axes[3, 0])
axes[3, 0].set_title('Model Size (MB)')

sns.barplot(x='Execution Time (s)', y='Model', hue='Method', data=combined_results, ax=axes[3, 1])
axes[3, 1].set_title('Execution Time (s)')

plt.tight_layout()
plt.show()
```

**Function used to predict the classification of a sequence using Phyicochemical properties**

```python
# Function to predict antimicrobial property using Method 1
def predict_method1(sequence, models, scaler):
    properties = calculate_physicochemical_properties(sequence)
    prop_values = scaler.transform([list(properties.values())])
    results = {}
    for name, model in models.items():
        prediction = model.predict(prop_values)
        results[name] = 'Antimicrobial' if prediction[0] == 1 else 'Non-antimicrobial'
    return results
```

```python
# Example prediction
sequence = 'AKKAGLWISEEKAL'

# Predictions using Method 1
predictions1 = predict_method1(sequence, trained_models_method1, scaler1)
display(HTML('<h2><b>Predictions for Method 1 (Physicochemical Properties):</b></h2>'))
for model_name, prediction in predictions1.items():
    display(HTML(f'<h2><b>{model_name}: {prediction}</b></h2>'))

```
